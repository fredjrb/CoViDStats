```python
from bs4 import BeautifulSoup
import json
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError
import pandas as pd
import numpy as np
```

```python
def gera_table_paises(url : str):
    headers ={'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    bytesHTML = ''
    try :
        request = Request(url,headers = headers)
        response = urlopen(request)
        bytesHTML = response.read()
    except HTTPError as e:
        print(e.status, e.reason)
    except URLError as e:    
        print(e.reason)

    strHTML = bytesHTML.decode('utf-8')    
    strHTML = " ".join(strHTML.split()).replace('> <', '><')
    soup = BeautifulSoup(strHTML, 'html.parser')
    table = soup.find('table', id='example2')
    return table
```

```python
def gera_dataframe_paises() :
    
    table = gera_table_paises('http://www.worldometers.info/geography/how-many-countries-are-there-in-the-world/')
    dictPaises = {}
    paises = table.findAll('a')

    for pais in paises :
        nome_pais = (pais.get_text().upper())
        area_pais = pais.find_parent('td').findNextSibling().findNextSibling().findNextSibling().get_text()
        dictPaises[nome_pais] = int(area_pais.replace(',', ''))

    dfArea = pd.DataFrame.from_dict(dictPaises, orient = 'index', columns=['AREA'])
    dfArea['PAIS'] = dfArea.index
    dfArea.index = range(dfArea.shape[0])
    return dfArea



def gera_dataframe_continente(url : str, continente : str) :
    table = gera_table_paises(url)
    dictPaises = {}
    paises = table.findAll('a')
    
    for pais in paises :
            nome_pais = (pais.get_text().upper())
            nome_continente = continente
            nome_regiao = pais.find_parent('td').findNextSibling().findNextSibling().get_text().upper()
            dictPaises[nome_pais] = [nome_continente, nome_regiao]
            
    dfArea = pd.DataFrame.from_dict(dictPaises, orient = 'index', columns=['CONTINENTE', 'SUB_REGIAO'])
    dfArea['PAIS'] = dfArea.index
    dfArea.index = range(dfArea.shape[0])
    return dfArea   


def gera_dataframe_africa() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-africa/', 'AFRICA')

def gera_dataframe_asia() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-asia/', 'ASIA')

def gera_dataframe_europa() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-europe/', 'EUROPA')

def gera_dataframe_am_latina() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-latin-america/', 'AMERICA LATINA')

def gera_dataframe_am_norte() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-northern-america/', 'AMERICA DO NORTE')

def gera_dataframe_oceania() :
    return gera_dataframe_continente('https://www.worldometers.info/geography/how-many-countries-in-oceania/', 'OCEANIA')
```

```python
def gera_codigo_pais() :
    table = gera_table_paises('https://www.worldometers.info/country-codes/')
    dictPaises = {}
    paises = table.findAll('a')
    for pais in paises :
        nome_pais = (pais.get_text().upper())
        iso_pais = pais.find_parent('td').findNextSibling().findNextSibling().findNextSibling().get_text()
        dictPaises[nome_pais] = iso_pais
    dfArea = pd.DataFrame.from_dict(dictPaises, orient = 'index', columns=['ISO'])
    dfArea['PAIS'] = dfArea.index
    dfArea.index = range(dfArea.shape[0])
    return dfArea    
```

```python
def gera_table_paises_regioes() :
    dfArea = pd.concat([gera_dataframe_africa(), gera_dataframe_asia(), gera_dataframe_europa(), gera_dataframe_am_latina(),
                        gera_dataframe_am_norte(), gera_dataframe_oceania()])
    dfArea.index = range(dfArea.shape[0])
    return dfArea

```

```python
dfRegioes = gera_table_paises_regioes()
dfISO = gera_codigo_pais()
dfArea = gera_dataframe_paises()
dfArea = pd.merge(dfArea, dfRegioes, how="inner", left_on = 'PAIS', right_on = 'PAIS')
dfArea = pd.merge(dfArea, dfISO, how="inner", left_on = 'PAIS', right_on = 'PAIS')
dfArea

```
