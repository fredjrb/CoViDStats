<h1>Raspagem dos dados de Piramide Populacional<h1>

### - Livro de Estatistica / Subir Notebook do Curso / Subir este notebook

```python
from bs4 import BeautifulSoup
import json
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError
import pandas as pd
import numpy as np
```

```python
def gera_json_faixa_etaria(strHTML : str, posMarcador : int) : 
    strHTML = " ".join(strHTML.split()).replace('> <', '><')
    soup = BeautifulSoup(strHTML, 'html.parser')
    scriptTag = soup.find('script', type="application/javascript")
    strJSon = scriptTag.string.split('[')[posMarcador].strip().split(']')[0].strip()
    strJSon = strJSon.replace("'v'", "'populacao'")
    strJSon = strJSon.replace("'k'", "'faixa_etaria'")
    strJSon = strJSon.replace("'", '"')
    strJSon = '[' + strJSon + ']'
    objeto = json.loads(strJSon)
    return objeto

def gera_dict_faixa_etaria(strHTML : str) : 
    male = gera_json_faixa_etaria(strHTML, 2)
    female = gera_json_faixa_etaria(strHTML, 3)
    faixa_etaria = {}
    for item in male :
        faixa_etaria[item['faixa_etaria']] = item['populacao']
    for item in female :
        faixa_etaria[item['faixa_etaria']] = item['populacao'] +  faixa_etaria[item['faixa_etaria']]
    return  faixa_etaria    

def get_localidade(strHTML : str) :
    strHTML = " ".join(strHTML.split()).replace('> <', '><')
    soup = BeautifulSoup(strHTML, 'html.parser')
    scriptTag = soup.find('script', type="application/javascript")
    strJSon = scriptTag.string.split('"')[1].strip().split('"')[0].strip()
    return strJSon

def gera_df_localidade(pais = None) :
    dataFrame = None
    if(pais is None) :
        pais = 'world'
    url = 'https://www.populationpyramid.net/'+ pais + '/2019/'
    headers ={'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    bytesHTML = ''
    try :
        request = Request(url,headers = headers)
        response = urlopen(request)
        bytesHTML = response.read()
    except HTTPError as e:
        print(e.status, e.reason)
    except URLError as e:    
        print(e.reason)
    dict_faixa = gera_dict_faixa_etaria(bytesHTML.decode('utf-8'))
    localidade = get_localidade(bytesHTML.decode('utf-8'))
    if(not localidade.isupper()) :
        dataFrame = pd.DataFrame.from_dict(data=dict_faixa, orient='index').T
        dataFrame.index = [localidade]
    return dataFrame

def get_lista_paises():
    url = 'https://www.populationpyramid.net/'
    headers ={'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    bytesHTML = ''
    try :
        request = Request(url,headers = headers)
        response = urlopen(request)
        bytesHTML = response.read()
    except HTTPError as e:
        print(e.status, e.reason)
    except URLError as e:    
        print(e.reason)
    strHTML = bytesHTML.decode('utf-8')    
    strHTML = " ".join(strHTML.split()).replace('> <', '><')
    soup = BeautifulSoup(strHTML, 'html.parser')
    scriptTag = soup.findAll('a', class_='countryLink')
    paises = [(tag['href'].split('/')[1]) for tag in scriptTag]
    return paises    
```

```python
popPyramidDF = pd.DataFrame()
paises = get_lista_paises()
paises = list(set(paises))
for pais in paises :
    dfPais = gera_df_localidade(pais)
    if dfPais is not None :
        popPyramidDF = popPyramidDF.append(dfPais)
popPyramidDF['country'] = popPyramidDF.index
popPyramidDF.country = popPyramidDF.country.str.upper()
popPyramidDF.index = range(popPyramidDF.shape[0])
popPyramidDF.to_csv('data/pyramid.csv', index=False, sep=',')
```
