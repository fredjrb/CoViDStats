## Avaliação Visual de Casos / Mortes por 100.000 (Histograma)

```python
def make_histograma(dados : pd.Series, descricao : str, ax : mplt.axes.Axes):
    #Usando a Regra de Sturges para definir o número de bins
    k = int(1 + 10/3 * np.log10(df.shape[0]))

    sns.set_style("whitegrid")
    ax = sns.histplot(data=dados, bins=k*5, palette=sns.color_palette("husl", n_colors=1), kde=True, ax = ax)
    ax.set_title(f'Distribuição de Frequência de {descricao}', fontsize = 18)
    ax.set_xlabel(descricao.upper(), fontsize=16)
    ax.set_ylabel('NÚMERO DE PAíSES', fontsize=16)
    ax.figure.set_size_inches(20,10)
    return ax
```

```python
fig = plt.figure(constrained_layout=True)
gs = GridSpec(2, 2, figure=fig)
fig.set_size_inches(20,40)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 0])


make_histograma(df.CASOS_POR_MILHAO, 'casos por milhão de habitantes', ax0)
make_histograma(df.MORTES_POR_MILHAO, 'mortes por milhão de habitantes', ax1)
make_histograma(df.LETALIDADE, 'letalidade', ax2)


```

### BOXPLOTS

```python
def make_bloxplot(dados : pd.Series, descricao : str, ax : mplt.axes.Axes) : 
    sns.set_palette(sns.diverging_palette(120, 0, l=25, n=1))
    ax = sns.boxplot(data=dados, orient='h', ax=ax)
    ax.figure.set_size_inches(20, 10)
    ax.set_title(f'Distribuição da Frequência de {descricao}' , fontsize=20)
    ax.set_xlabel(descricao.upper(), fontsize=14)
    ax.set_xlim(xmin=0, xmax=(dados.max() * 1.10))
    ax.grid(True)
    return ax

def outliers(series : pd.Series) :
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iiq = (q3 - q1)
    inf = q1 - (1.5 * iiq)
    sup = q3 + (1.5 * iiq)
    if (inf < 0) :
        inf = 0
    return inf, sup
```

```python
fig = plt.figure(constrained_layout=True)
gs = GridSpec(2, 2, figure=fig)
fig.set_size_inches(40,40)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 0])

make_bloxplot(df.CASOS_POR_MILHAO, 'Casos por Milhão de Habitantes', ax0)
make_bloxplot(df.MORTES_POR_MILHAO, 'Mortes por Milhão de Habitantes', ax1)
make_bloxplot(df.LETALIDADE, 'Letalidade', ax2)
```

```python
_, sup_casos = outliers(df.CASOS_POR_MILHAO)

print(f'Considera-se outliers no número de casos, países com mais de {sup_casos} casos por milhão habitantes')
paisesOutliersCasos =  pd.DataFrame()
paisesOutliersCasos = df.query(f'CASOS_POR_MILHAO > {sup_casos}')[['PAIS', 'CASOS_POR_MILHAO', 'CONTINENTE', 'SUB_REGIAO']]
paisesOutliersCasos.columns = ['PAIS', 'CASOS POR MILHAO', 'CONTINENTE', 'SUB REGIAO']
paisesOutliersCasos
```

```python
def box_plot_continente(info : str, ax : mplt.axes.Axes):
    sns.set_style("whitegrid")
    classes = df.CONTINENTE.unique()
    classes.sort()
    sns.boxplot(x='CONTINENTE', data=df, orient='v', y=info, order=classes,  palette=sns.color_palette("husl", n_colors=30), ax=ax)    
    ax.set_title(info, fontsize=20)
    ax.figure.set_size_inches(20,20) 
    ax.figure.autofmt_xdate(rotation=45)
    return ax
```

```python
fig = plt.figure(constrained_layout=False)
gs = GridSpec(2, 2, figure=fig)
fig.set_size_inches(40,40)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 0])


box_plot_continente('CASOS_POR_MILHAO', ax0)
box_plot_continente('MORTES_POR_MILHAO', ax1)
box_plot_continente('LETALIDADE', ax2)
```

```python
def box_plot_sub_regiao(info : str, ax : mplt.axes.Axes):
    sns.set_style("whitegrid")
    classes = df.SUB_REGIAO.unique()
    classes.sort()
    sns.boxplot(x='SUB_REGIAO', data=df, orient='v', y=info, order=classes,  palette=sns.color_palette("husl", n_colors=30), ax=ax)    
    ax.set_title(info, fontsize=20)
    ax.figure.set_size_inches(20,20) 
    ax.figure.autofmt_xdate(rotation=45)
    return ax
```

```python
fig = plt.figure(constrained_layout=False)
gs = GridSpec(2, 2, figure=fig)
fig.set_size_inches(40,40)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 0])


box_plot_sub_regiao('CASOS_POR_MILHAO', ax0)
box_plot_sub_regiao('MORTES_POR_MILHAO', ax1)
box_plot_sub_regiao('LETALIDADE', ax2)
```

### Buscando Outliers no número de Casos e Mortes por 100 mil Habitantes

```python
def linha_tendencia(x : pd.Series, y : pd.Series) :
    X = sm.add_constant(x)
    results = sm.OLS(y, X).fit()
    b = results.params[0]
    a = results.params[1]
    resultados = []
    for i in range(x.size) :
        if(not np.isnan(x[i])) :
            resultados.append(a * x[i] + b)
    return resultados   
```

```python
df.CASOS_POR_MILHAO.isna().value_counts()
```

```python
df.index = range(df.shape[0])
resultados = linha_tendencia(df.CASOS_POR_MILHAO, df.MORTES_POR_MILHAO)
```
