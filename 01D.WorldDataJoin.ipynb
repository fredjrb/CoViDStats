```python
data_ref= '2021-07-19'

def map_value(serie : pd.Series, mapValues : dict):
    newSerie = []
    for value in serie :
        newValue = mapValues.get(value, value)
        newSerie.append(newValue)
    return pd.Series(newSerie)

# Dados do site de Index Mundo
dfIndexMundi = pd.read_csv('data/pop_index_mundi.csv', sep = ',') 

# Dados do Worldometers (https://www.worldometers.info/)
dfArea = pd.read_csv('data/area.csv', sep = ',')

# Dados do CIA World Fact 
dfCIA = pd.read_csv('data/data_cia.csv', sep = ',')
```

### Preprocessamento do dataframe de Piramide Populacional

```python
grupos_etarios = ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39',
                  '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79',
                  '80-84', '85-89', '90-94', '95-99', '100+']

dfIndexMundi['POPULACAO'] = dfIndexMundi[grupos_etarios].sum(axis=1)
dfIndexMundi.head()   
```

### Montagem da base geográfica básica

```python
dfArea.PAIS = map_value(dfArea.PAIS, mapDFArea)
dfIndexMundi.PAIS = map_value(dfIndexMundi.PAIS, mapDFGeo) 
dfCIA.PAIS = map_value(dfCIA.PAIS, mapDFGeo)

dfGeo = pd.merge(dfArea, dfIndexMundi, how="left", left_on = 'PAIS', right_on = 'PAIS')
dfGeo = pd.merge(dfGeo, dfCIA, how="left", left_on = 'PAIS', right_on = 'PAIS')


dfGeo.head()
```

### Preprocessamento do dataframe do WHO

```python
# Dados do WHO
dfWHOCovid = pd.read_csv('data/covid_who.csv', sep = ',') 
dfWHOCovid.Name = dfWHOCovid.Name.str.upper()
dfWHOCovid.Name = map_value(dfWHOCovid.Name, mapDFGeo)


columnsToDrop = ['WHO Region',
                 'Cases - newly reported in last 7 days',
                 'Cases - newly reported in last 7 days per 100000 population',
                 'Cases - newly reported in last 24 hours',
                 'Deaths - newly reported in last 7 days',
                 'Deaths - newly reported in last 7 days per 100000 population',
                 'Deaths - newly reported in last 24 hours',
                 'Cases - cumulative total per 100000 population',
                 'Deaths - cumulative total per 100000 population',
                 'Transmission Classification']

dfWHOCovid['CASOS_POR_MILHAO'] = dfWHOCovid['Cases - cumulative total per 100000 population'] * 10
dfWHOCovid['MORTES_POR_MILHAO'] = dfWHOCovid['Deaths - cumulative total per 100000 population'] * 10

dfWHOCovid.drop(columns=columnsToDrop, inplace=True)

dfWHOCovid.columns = ['PAIS', 'CASOS', 'MORTES', 'CASOS_POR_MILHAO', 'MORTES_POR_MILHAO']
dfWHOCovid.head()
```

### Usando a Base da WHO para dados de COVID

```python
df = pd.merge(dfGeo, dfWHOCovid, how="left", left_on = 'PAIS', right_on = 'PAIS')
df.to_csv('data/consolidadoWHO.csv', index=False, sep=',')
df.shape
```

### Preprocessamento do dataframe do OWID

```python
dfOWIDCovid = pd.read_csv('data/covid_owid.csv', sep = ',')
dfOWIDCovid = dfOWIDCovid.query(f'date=="{data_ref}"').copy()
dfOWIDCovid.index = range(dfOWIDCovid.shape[0])

dfOWIDCovid.columns

dfOWIDCovid.drop(labels=['iso_code', 'continent', 'new_cases_smoothed', 'new_cases', 'new_deaths', 'new_deaths_smoothed', 'new_cases_per_million', 
                    'new_cases_smoothed_per_million','new_deaths_per_million', 'new_deaths_smoothed_per_million', 
                    'weekly_icu_admissions', 'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
                    'weekly_hosp_admissions_per_million', 'new_tests', 'new_tests_per_thousand',
                    'new_tests_smoothed', 'new_tests_smoothed_per_thousand', 'positive_rate', 'tests_per_case', 
                    'tests_units', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations',
                    'new_vaccinations_smoothed', 'total_vaccinations_per_hundred',
                    'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred',
                    'new_vaccinations_smoothed_per_million', 'stringency_index', 'median_age', 'aged_65_older',
                    'aged_70_older', 'icu_patients', 'icu_patients_per_million',
                    'hosp_patients', 'hosp_patients_per_million', 'total_tests',
                    'total_tests_per_thousand', 'population', 'population_density', 
                    'life_expectancy', 'excess_mortality'], axis='columns', inplace=True)


dfOWIDCovid.location = dfOWIDCovid.location.str.upper() 

dfOWIDCovid.location = map_value(dfOWIDCovid.location, mapDFGeo)

dfOWIDCovid.columns = ['PAIS', 'DATA', 'CASOS', 'MORTES', 'CASOS_POR_MILHAO','MORTES_POR_MILHAO',
                       'FATOR R', 'PIB_PER_CAPITA', 'EXTREMA_POBREZA', 'TAXA_MORTES_CARDIOVASCULAR', 'PREVALENCIA_DIABETES', 
                       'MULHERES_FUMANTES', 'HOMENS_FUMANTES', 'ESTRUTURA_LAVAGEM_MAOS', 'LEITOS_HOSPITALARES_POR_MIL', 
                       'IDH']

dfOWIDCovid.head()
```

### Usando a Base do OWID para dados de COVID

```python
df2 = pd.merge(dfGeo, dfOWIDCovid, how="left", left_on = 'PAIS', right_on = 'PAIS')
df2.to_csv('data/consolidadoOWID.csv', index=False, sep=',')
df2.shape
```
